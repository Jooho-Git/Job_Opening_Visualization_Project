## 채용공고 textrank 키워드

```python
from konlpy.tag import Mecab 
mecab = Mecab(dicpath='C:/mecab/mecab-ko-dic')
```

```python
tag_classes = ['NNG', 'NNP','VA', 'VV+EC','VV+ETM', 'NP', 'NNBC', 'VA+EC','SL']

def mecab_tokenizer(text):   
    clean_word = []
    result = mecab.pos(text, join=True)
    for i in result:
        word = i.split('/')[0]
        tag = i.split('/')[1]
        if tag in tag_classes:
            if (word not in stopword) and len(word) > 1:
                clean_word.append(word)
    return clean_word   
```

```python
# 통합_2021
from textrank import KeywordSummarizer

summarizer = KeywordSummarizer(tokenize=mecab_tokenizer, min_count=2, min_cooccurrence=1)
textrank_2021 = summarizer.summarize(total_2021, topk=30)
```

```python
# 통합_2021 제외
from textrank import KeywordSummarizer

summarizer = KeywordSummarizer(tokenize=mecab_tokenizer, min_count=2, min_cooccurrence=1)
textrank_else = summarizer.summarize(total_else, topk=30)
```

```python
keyword = []
score = []

for k in textrank_2021:
    keyword.append(k[0])
    score.append(k[1])    
for k in textrank_else:
    keyword.append(k[0])
    score.append(k[1])
    
total_data = pd.DataFrame()
total_data['직무명'] = ['데이터 분석가' for _ in range(60)]
total_data['년도'] = ['2021' for _ in range(30)] + ['2021이전' for _ in range(30)]
total_data['키워드'] = pd.Series(keyword)
total_data['빈도'] = pd.Series(score)
```

```python
total_data.to_excel('채용 키워드.xlsx')
```
