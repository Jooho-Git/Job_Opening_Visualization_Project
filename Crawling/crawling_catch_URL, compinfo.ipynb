```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
# 드라이버 실행
driver = webdriver.Chrome()
```

```python
# 기업 리스트 가져오기 & reshaping

jobplanet_excel = pd.read_excel("잡플래닛_기업리스트.xlsx", index_col = 0)

jobplanet_reshape = pd.melt(jobplanet_excel)
jobplanet_reshape.head()
```

## 캐치 URL 수집

```python
# url 수집

urls_2 = []

html = driver.page_source # 페이지 소스 저장
html = BeautifulSoup(html, 'html.parser')

for i in jobplanet_reshape.iloc[444:,1]:
    # 접속하기
    driver.get("https://www.catch.co.kr/Comp/CompMajor?flag=Search")
    time.sleep(3)
    
    # 검색하기
    search = driver.find_element_by_xpath('//*[@id="CompName"]')
    search.send_keys(i)
    time.sleep(1)
    
    driver.find_element_by_xpath('//*[@id="imgSearch"]').click()
    time.sleep(3)
    
    # url 수집
    try:
        url=driver.find_element_by_xpath('//*[@id="updates"]/tbody/tr[1]/td[1]/dl/dt[2]/a').get_attribute('href')
        urls_2.append(url)
    except:
        try: 
            url=driver.find_element_by_xpath('//*[@id="updates"]/tbody/tr/td[1]/dl/dt[2]/a').get_attribute('href')
            urls_2.append(url)
        except:
            urls_2.append(i)
    
print(urls_2)
```

```python
import csv

with open("urls_2.csv", 'w') as file:
    writer = csv.writer(file)
    writer.writerow([i for i in range(0, 556)])
    writer.writerow(urls_2)
```
